**11. Investigate the table of modal distributions and look for other patterns. Try to explain them in terms of your own impressionistic understanding of the different genres. Can you find other closed classes of words that exhibit significant differences across different genres?**

```python
from nltk.corpus import brown
from collections import Counter
import matplotlib.pyplot as plt

# Ensure NLTK corpora are downloaded
nltk.download('brown')

# List of modal verbs
modals = ['can', 'could', 'may', 'might', 'must', 'will']

# Closed classes of words: Pronouns, prepositions, conjunctions
pronouns = ['i', 'you', 'he', 'she', 'it', 'we', 'they']
prepositions = ['in', 'on', 'at', 'by', 'with', 'about', 'against', 'between', 'into']
conjunctions = ['and', 'or', 'but', 'if', 'because', 'while', 'although']

# Function to count modal verbs and other closed classes in a genre
def count_words_in_genre(genre, word_list):
    words = brown.words(categories=genre)
    words_lower = [word.lower() for word in words]
    word_counts = Counter(words_lower)
    return {word: word_counts[word] for word in word_list}

# Genres to analyze
genres = ['news', 'hobbies', 'fiction', 'government', 'religion']

# Store modal counts by genre
modal_distributions = {genre: count_words_in_genre(genre, modals) for genre in genres}
pronoun_distributions = {genre: count_words_in_genre(genre, pronouns) for genre in genres}
preposition_distributions = {genre: count_words_in_genre(genre, prepositions) for genre in genres}
conjunction_distributions = {genre: count_words_in_genre(genre, conjunctions) for genre in genres}

# Function to plot word distribution
def plot_distributions(word_distributions, title):
    for genre, counts in word_distributions.items():
        words, frequencies = zip(*sorted(counts.items()))
        plt.plot(words, frequencies, label=genre)
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Plotting modal distributions
plot_distributions(modal_distributions, "Modal Verb Distributions Across Genres")

# Plotting other closed class distributions
plot_distributions(pronoun_distributions, "Pronoun Distributions Across Genres")
plot_distributions(preposition_distributions, "Preposition Distributions Across Genres")
plot_distributions(conjunction_distributions, "Conjunction Distributions Across Genres")
```

**16. Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?**

```python
# Generate table with tokens, types and lexical diversity for each genre
table = []
for genre in brown.categories():
    words = brown.words(categories=genre)
    words_lower = [word.lower() for word in words]
    num_tokens = len(words)
    num_types = len(set(words))
    lex_diversity = num_types / num_tokens
    table.append((genre, num_tokens, num_types, lex_diversity))

# Print table
print("Genre\t\tTokens\tTypes\tLexical Diversity")
for row in table:
    genre, num_tokens, num_types, lex_diversity = row
    # shortten the genre name
    genre = genre[:6]
    print(f"{genre}\t\t{num_tokens}\t{num_types}\t{lex_diversity:.4f}")
```

**17. Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.**

```python
from nltk.corpus import stopwords

def most_frequent_words(text):
    words = [word.lower() for word in text if word.isalpha()]
    word_counts = Counter(words)
    stop_words = set(stopwords.words('english'))
    for stop_word in stop_words:
        del word_counts[stop_word]
    return word_counts.most_common(50)
```

```python
print(most_frequent_words(brown.words(categories='news')))
```

**18. Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.**

```python
# I believe punctuation should be removed too
def most_frequent_bigrams(text):
    bigrams = nltk.bigrams(text)
    bigram_counts = Counter(bigrams)
    stop_words = set(stopwords.words('english'))
    with_stopwords = []
    # Collect bigrams that contain stopwords
    with_stopwords = [bigram for bigram in bigram_counts if bigram[0].lower() in stop_words or bigram[1].lower() in stop_words]
    
    # Remove bigrams with stopwords
    for bigram in with_stopwords:
        del bigram_counts[bigram]
    return bigram_counts.most_common(50)
```

```python
print(most_frequent_bigrams(brown.words(categories='news')))
```

**23. Zipf's Law: Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f × r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.**
1. Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?
2. Generate random text, e.g., using random.choice("abcdefg "), taking care to include the space character. You will need to import random first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, and generate the Zipf plot as before, and compare the two plots. What do you make of Zipf's Law in the light of this?

```python
import nltk
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import random

# Download the necessary NLTK data
# nltk.download('gutenberg')
# nltk.download('punkt')

# Function to process a large text and plot word frequency vs. rank
def zipf_plot(text):
    # Tokenize the text
    words = nltk.word_tokenize(text.lower())
    
    # Count word frequencies
    word_counts = Counter(words)
    
    # Sort words by frequency
    sorted_word_counts = sorted(word_counts.values(), reverse=True)
    
    # Rank the words
    ranks = np.arange(1, len(sorted_word_counts) + 1)
    
    # Plot on a log-log scale
    plt.figure(figsize=(10, 6))
    plt.loglog(ranks, sorted_word_counts, marker="o", linestyle="none", label="Word Frequency")
    plt.xlabel("Rank (log scale)")
    plt.ylabel("Frequency (log scale)")
    plt.title("Word Frequency vs. Rank (Zipf's Law)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Use a large text from NLTK's Gutenberg corpus (e.g., 'Moby Dick')
from nltk.corpus import gutenberg
text = gutenberg.raw('melville-moby_dick.txt')

# Call the function
zipf_plot(text)
```

```python
import string

# Function to generate random text and plot the Zipf distribution
def random_text_zipf_plot(length=100000):
    # Generate random text using 'abcdefg ' (space included to create "words")
    chars = list("abcdefg ")
    random_text = ''.join(random.choice(chars) for _ in range(length))
    
    # Tokenize by splitting on space (since this creates "words")
    words = random_text.split()
    
    # Count word frequencies
    word_counts = Counter(words)
    
    # Sort word counts and generate ranks
    sorted_word_counts = sorted(word_counts.values(), reverse=True)
    ranks = np.arange(1, len(sorted_word_counts) + 1)
    
    # Plot on a log-log scale
    plt.figure(figsize=(10, 6))
    plt.loglog(ranks, sorted_word_counts, marker="o", linestyle="none", color="red", label="Random Text Frequency")
    plt.xlabel("Rank (log scale)")
    plt.ylabel("Frequency (log scale)")
    plt.title("Word Frequency vs. Rank (Random Text)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate random text and plot
random_text_zipf_plot(100000)
```

Zipf's Law connects to a key property of Barabási-Albert (BA) networks, specifically the scale-free nature of these networks. Both describe phenomena where a small number of nodes (or words, in Zipf's case) dominate in terms of connections or occurrences. In BA networks, the degree distribution follows a power-law: most nodes have few connections, while a few highly connected nodes (hubs) emerge. This mirrors the distribution in Zipf's Law, where a few words appear very frequently, and most words are rare. The preferential attachment mechanism in BA networks is analogous to how common words in language are used more frequently, reinforcing their prominence, just as highly connected nodes in the network continue to accumulate connections.

### Other Important Power-Laws on the Internet:

**Webpage Connectivity:** The structure of the web, where some websites (e.g., Google, Wikipedia) receive a disproportionate number of links compared to the vast majority of websites, follows a power-law distribution. This creates a small-world network where a few websites act as hubs with a vast number of connections, while most sites have relatively few inbound links.

**Social Media Networks:** Platforms like Facebook, Twitter, and Instagram also exhibit power-law distributions in terms of connections (followers or friends). A small number of users, often celebrities or influencers, have millions of followers, while the majority have far fewer connections. This creates a scale-free network, with a significant influence concentrated in the hands of a few.

**Internet Traffic:** The distribution of data traffic across websites follows a power-law as well. A handful of popular websites like YouTube or Netflix account for a large percentage of global internet traffic, while the vast majority of websites see much lower traffic. This power-law distribution in internet usage is a key factor in the design and scaling of internet infrastructure.

These instances of power-laws reveal that a small number of entities or nodes can dominate networks, leading to highly unequal distributions, which is a common theme across many types of complex systems.
